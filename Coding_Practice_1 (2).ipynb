### Initial Setup
Execute the below code cell for initial setup

````python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset using seaborn
data = sns.load_dataset('iris')

# Display the first few rows of the dataset
data.head()
````

Q1. What are the minimum and maximum values of petal_width for each species?


````python
# Calculate the min and max values of petal_width for each species
# Write the code here
petal_width_min_max = data.groupby('species')['petal_width'].agg(['min', 'max'])

print(petal_width_min_max)
````

Q2. For each species, calculate the range (maximum - minimum) of the sepal_width

````python
# Calculate the range of sepal_width for each species
# Write the code here
sepal_width_range = data.groupby('species')['sepal_width'].agg(['min', 'max'])
sepal_width_range['range'] = sepal_width_range['max'] - sepal_width_range['min']

print(sepal_width_range[['range']])
````

Q3. Calculate the average sepal length for each species.  Print the result as a Series.

---



````python
# Write the code here
avg_sepal_length = data.groupby('species')['sepal_length'].mean()

print(avg_sepal_length)
````

Q4. Count the number of samples with sepal length > 5.0 cm

````python
# Write the code here
sepal_length_above_5 = data[data['sepal_length'] > 5.0].shape[0]

print(sepal_length_above_5)
````

Q5. Determine which species have average sepal width > 2.5 cm in desc order. Print the result as a Series.

````python
# Write the code here
avg_sepal_width = data.groupby('species')['sepal_width'].mean().sort_values(ascending=False)
species_above_2cm = avg_sepal_width[avg_sepal_width > 2.5]
print(species_above_2cm)
````

Q6. Compute the standard deviation of sepal length for 'virginica'. Print the result rounded to 15 decimal places followed by ' cm'.

(Example format: 0.123456789012345 cm)

````python
# Write the code here
virginica_sepal_length_std = data[data['species'] == 'virginica']['sepal_length'].std()
print(f"{virginica_sepal_length_std:.15f} cm")
````

Q7. Calculate and print the average petal length for each species in the given format. Then create a bar plot showing the average petal length for each species with the title 'Average Petal Length by Species.

```
      species  petal_length
0      setosa         -
1  versicolor         -
2   virginica         -
```

````python
# Write the code here
avg_petal_len = data.groupby('species')['petal_length'].mean().reset_index()
print(avg_petal_len)
````

````python
# Write the code here to create visual
sns.barplot(data=avg_petal_len, x='species', y='petal_length')
plt.title('Average Petal Length by Species')
plt.ylabel('Petal Length')
plt.show()
````

Q8. Find and print the min and max sepal width for each species in the given format. Then create a grouped bar plot showing the min and max sepal width for each species with the title 'Min and Max Sepal Width by Species.

```
      species  min  max
0      setosa   -    -
1  versicolor   -    -
2   virginica   -    -
```

````python
# Write the code here
min_max_sw = data.groupby('species')['sepal_width'].agg(['min', 'max']).reset_index()
print(min_max_sw)
````

````python
# Write the code here to create visual
min_max_sw_melted = min_max_sw.melt(id_vars='species', var_name='Type', value_name='Sepal Width')
sns.barplot(data=min_max_sw_melted, x='species', y='Sepal Width', hue='Type')
plt.title('Min and Max Sepal Width by Species')
plt.show()
````

Q9. Group data by species and calculate the standard deviation of petal width for each species in the given format. Then create a bar plot showing the standard deviation of petal width for each species with the title 'Standard Deviation of Petal Width by Species'.

```
      species  petal_width
0      setosa     -
1  versicolor     -
2   virginica     -
```

````python
# Write the code here
std_petal_width = data.groupby('species')['petal_width'].std().reset_index()
print(std_petal_width)
````

````python
# Write the code here to create visual
sns.barplot(data=std_petal_width, x='species', y='petal_width')
plt.title('Standard Deviation of Petal Width by Species')
plt.ylabel('Petal Width Std Dev')
plt.show()
````

Q10. Find and print the correlation between sepal length and petal length in the given format. Then create a scatter plot with a regression line showing the relationship between sepal length and petal length.

- Format:

```
   Correlation
0     -
```

````python
# Write the code here
correlation = data[['sepal_length', 'petal_length']].corr().iloc[0, 1]
corr = pd.DataFrame({'Correlation': [correlation]})
print(corr)
````

````python
# Write the code here to create visual
sns.regplot(data=data, x='sepal_length', y='petal_length')
plt.title(f'Sepal vs Petal Length (Correlation = {correlation:.2f})')
plt.show()
````
